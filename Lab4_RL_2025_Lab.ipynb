Импортируем библиотеки, создаем свою среду размера 6х6. S обозначает точку старта. F -- лед безопасен, H -- проталина, G -- цель. Параметр `is_slippery=False` отвечает за условное отсутствие скольжения. То есть если агент выбрал действие пойти направо, то он переместится в соответствующее состояние.
````
import gym
import numpy as np
import time
from IPython.display import clear_output


def generate_random_map(size, p, sd):
    """Generates a random valid map (one that has a path from start to goal)
    :param size: size of each side of the grid
    :param p: probability that a tile is frozen
    """
    valid = False
    np.random.seed(sd)

    # DFS to check that it's a valid path.
    def is_valid(res):
        frontier, discovered = [], set()
        frontier.append((0,0))
        while frontier:
            r, c = frontier.pop()
            if not (r,c) in discovered:
                discovered.add((r,c))
                directions = [(1, 0), (0, 1), (-1, 0), (0, -1)]
                for x, y in directions:
                    r_new = r + x
                    c_new = c + y
                    if r_new < 0 or r_new >= size or c_new < 0 or c_new >= size:
                        continue
                    if res[r_new][c_new] == 'G':
                        return True
                    if (res[r_new][c_new] not in '#H'):
                        frontier.append((r_new, c_new))
        return False

    while not valid:
        p = min(1, p)
        res = np.random.choice(['F', 'H'], (size, size), p=[p, 1-p])
        res[0][0] = 'S'
        res[-1][-1] = 'G'
        valid = is_valid(res)
    return ["".join(x) for x in res]

#Генерация карты
random_map = generate_random_map(size=6, p=0.8, sd = random_seed) #Создаем свою карту
env = gym.make("FrozenLake-v0", desc=random_map, is_slippery=False) #Инициализируем среду
print("Ваша карта")
env.render() #Выводим карту на экран
````

# Задача 2
Дополните следующий код так, чтобы в результате обучения модели можно было узнать количество побед и номер игры (`game`), на котором агент впервые одержал пятую победу подряд.

Поясним, что возвращает функция ```env.step(action)```

```state2``` -- следующее состояние

```reward``` -- награда

```done``` -- флаг окончания игры. True в случае победы или падения в проталину. False в остальных случаях.




````
from tqdm import tqdm

# Инициализация
np.random.seed(random_seed)
total_games = 10000
max_steps = 100
Q = np.zeros((env.observation_space.n, env.action_space.n))

# Переменные для подсчета побед
wins = 0
consecutive_wins = 0
first_five_wins_game = None

# Основной цикл
for game in tqdm(range(total_games)):
    state = env.reset()
    t = 0
    game_won = False

    while t < max_steps:
        t += 1

        action = choose_action(state)

        state2, reward, done, info = env.step(action)

        if t == max_steps:
            done = True

        learn(state, state2, reward, action, done)

        state = state2

        if done:
            if reward == 1.0:
                wins += 1
                consecutive_wins += 1
                game_won = True
            else:
                consecutive_wins = 0
            break

    # Проверка: 5 побед подряд впервые
    if first_five_wins_game is None and consecutive_wins == 5:
        first_five_wins_game = game + 1  # +1, т.к. нумерация с нуля
````

Кроме того, у функции `learn()` должен появиться дополнительный аргумент (следующее действие). Ниже приведен фрагмент кода с пояснениями, как именно нужно модифицировать алгоритм.



````
def learn(state, state2, reward, action, action2, done):
    if done:
        Q[state, action] = Q[state, action] + lr_rate * (reward - Q[state, action])
    else:
        Q[state, action] = Q[state, action] + lr_rate * (reward + gamma * Q[state2, action2] - Q[state, action])
````

````
from tqdm import tqdm

np.random.seed(random_seed)
total_games = 40000
max_steps = 100
Q = np.zeros((env.observation_space.n, env.action_space.n))

wins = 0
consecutive_wins = 0
first_five_wins_game = None

for game in tqdm(range(total_games)):
    state = env.reset()
    t = 0
    action = choose_action(state)  # Выбор первого действия

    while t < max_steps:
        t += 1

        state2, reward, done, info = env.step(action)
        action2 = choose_action(state2)  # Выбор следующего действия

        if t == max_steps:
            done = True

        learn(state, state2, reward, action, action2, done)

        state = state2
        action = action2

        if done:
            if reward == 1.0:
                wins += 1
                consecutive_wins += 1
            else:
                consecutive_wins = 0

            if first_five_wins_game is None and consecutive_wins == 5:
                first_five_wins_game = game + 1  # игры нумеруются с 0

            break

````

````
# Реализация других стратегий
random_seed = 100
env = gym.make("FrozenLake-v0", desc=generate_random_map(size=6, p=0.8, sd=random_seed), is_slippery=False)
np.random.seed(random_seed)

def train_agent(strategy, total_games=10000, epsilon=0.1, tau=1.0, c=1.0, init_q=0.0, lr_rate=0.1, gamma=0.8):
    Q = np.full((env.observation_space.n, env.action_space.n), init_q, dtype=np.float32)
    N = np.zeros_like(Q, dtype=np.int32)  # Для UCB
    rewards = []

    for game in tqdm(range(total_games)):
        state = env.reset()
        total_reward = 0

        for _ in range(100):  # Макс. шагов на эпизод
            # --- Выбор действия ---
            if strategy == "epsilon_greedy":
                if np.random.rand() < epsilon:
                    action = np.random.randint(env.action_space.n)
                else:
                    action = np.random.choice(np.flatnonzero(Q[state] == Q[state].max()))

            elif strategy == "softmax":
                preferences = Q[state] / tau
                preferences -= np.max(preferences)  # числовая стабильность
                exp_preferences = np.exp(preferences)
                probs = exp_preferences / np.sum(exp_preferences)
                action = np.random.choice(env.action_space.n, p=probs)

            elif strategy == "ucb":
                total_visits = np.sum(N[state]) + 1e-5
                confidence = c * np.sqrt(np.log(total_visits) / (N[state] + 1e-5))
                ucb_values = Q[state] + confidence
                action = np.argmax(ucb_values)

            elif strategy == "optimistic":
                action = np.random.choice(np.flatnonzero(Q[state] == Q[state].max()))

            else:
                raise ValueError("Неизвестная стратегия!")

            # --- Шаг в среде ---
            state2, reward, done, _ = env.step(action)
            total_reward += reward

            # --- Обновление Q-таблицы ---
            Q[state, action] += lr_rate * (reward + gamma * np.max(Q[state2]) - Q[state, action])
            if strategy == "ucb":
                N[state, action] += 1

            state = state2
            if done:
                break

        rewards.append(total_reward)

    return np.cumsum(rewards) / (np.arange(len(rewards)) + 1)
````

````
# Настройки стратегий
strategies = {
    "ε-Greedy": {"strategy": "epsilon_greedy", "epsilon": 0.1},
    "Softmax": {"strategy": "softmax", "tau": 0.1},
    "UCB": {"strategy": "ucb", "c": 2.0},
    "Optimistic": {"strategy": "optimistic", "init_q": 5.0},
}
````

````
# Обучение
results = {}
for name, params in strategies.items():
    results[name] = train_agent(**params, total_games=10000)
````

````
# Визуализация
plt.figure(figsize=(12, 6))
for name, performance in results.items():
    plt.plot(performance, label=name)
plt.title("Средняя награда за эпизод")
plt.xlabel("Эпизод")
plt.ylabel("Средняя награда")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
````

````
# влияние параметров
def run_sarsa(gamma=0.9, lr_rate=0.1, epsilon=0.1, total_games=40000, max_steps=100):
    Q = np.zeros((env.observation_space.n, env.action_space.n))
    total_wins = 0

    def choose_action(state):
        if np.random.rand() < epsilon:
            return np.random.randint(env.action_space.n)
        else:
            return np.random.choice(np.flatnonzero(Q[state] == Q[state].max()))

    def learn(state, action, reward, state2, action2, done):
        target = reward if done else reward + gamma * Q[state2, action2]
        Q[state, action] += lr_rate * (target - Q[state, action])

    for _ in range(total_games):
        state = env.reset()
        action = choose_action(state)

        for _ in range(max_steps):
            state2, reward, done, _ = env.step(action)
            action2 = choose_action(state2)
            learn(state, action, reward, state2, action2, done)

            state, action = state2, action2

            if done:
                if reward == 1:
                    total_wins += 1
                break

    return total_wins

# --- Гиперпараметры для исследования ---
x_values = np.linspace(0.01, 1.0, 10)
gamma_results = []
lr_results = []
epsilon_results = []

# Прогон экспериментов
for val in tqdm(x_values, desc="Запуск экспериментов"):
    gamma_results.append(run_sarsa(gamma=val, lr_rate=0.1, epsilon=0.1))
    lr_results.append(run_sarsa(gamma=0.9, lr_rate=val, epsilon=0.1))
    epsilon_results.append(run_sarsa(gamma=0.9, lr_rate=0.1, epsilon=val))

# Визуализация результатов
plt.figure(figsize=(10, 6))
plt.plot(x_values, gamma_results, 'o-', label='Gamma (γ)', color='blue')
plt.plot(x_values, lr_results, 's-', label='Learning Rate (α)', color='green')
plt.plot(x_values, epsilon_results, '^-', label='Epsilon (ε)', color='red')

plt.xlabel("Значение гиперпараметра")
plt.ylabel("Количество побед за 40 000 игр")
plt.title("Влияние γ, α, ε на производительность SARSA")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
````
